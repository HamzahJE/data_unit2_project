### Step 3: Treat Missing Values

```python
######################### TO DO ###################################

####### FOLLOW PROJECT DIRECTIONS TO COMPLETE THIS STEP ###########
```

```python
### DEFINE ###
#since we only have a few (less than 10) null values we can just drop them.

### CLEAN ### 
house_data.dropna(inplace=True)
### TEST ###
house_data.info()
```

```python
### DEFINE ###
#since we only have a few (less than 10) null values we can just drop them.

### CLEAN ### 
sales_data.dropna(inplace=True)
### TEST ###
sales_data.info()
```

```python
### DEFINE ###
#drop unnamed: 0 column from house_data

### CLEAN ### 
house_data.drop(columns='unnamed: 0',inplace=True)

### TEST ###
house_data.head()
```

```python
### DEFINE ###
#drop unnamed: 0 column from sales_data

### CLEAN ### 
sales_data.drop(columns='unnamed: 0',inplace=True)

### TEST ###
sales_data.head()
```

```python
### DEFINE ###
#dropping the year_renovated column

### CLEAN ### 
house_data.isna().sum()
house_data.drop(columns='year_renovated',inplace=True)

### TEST ###
house_data.head()
```

```python
sales_data.head()
```

```python
### DEFINE ###
# changing price to float and removing ',' and '$'
#sales id type change to string 
### CLEAN ### 
sales_data.price=sales_data.price.str.replace('$','')
sales_data.price=sales_data.price.str.replace(',','')
sales_data.price=sales_data.price.astype('float')
sales_data.id=sales_data.id.astype(str)

### TEST ###
sales_data.head()
```

```python
### DEFINE ###
#converting to datetime and removing the T00000 to do so 
### CLEAN ### 
sales_data.date=sales_data.date.str.replace('T000000','')
sales_data.date=pd.to_datetime(sales_data.date)
sales_data.date =sales_data.date.dt.strftime('%m/%d/%Y')

### TEST ###
sales_data.head()
```

```python
### DEFINE ###
#house_data datatype fix

### CLEAN ### 
house_data.grade=house_data.grade.fillna(house_data.grade.mode()[0])
house_data.grade=pd.to_numeric(house_data.grade,errors='coerce').astype('int')
house_data.year_built=pd.to_numeric(house_data.year_built,errors='coerce').astype(int)

house_data.id=house_data.id.astype(str)
house_data.zipcode=house_data.zipcode.astype(str)
house_data.long=house_data.long.astype(str)
house_data.lat=house_data.lat.astype(str)
house_data.floors=house_data.floors.astype('int')
house_data.grade=house_data.grade.astype('int')
house_data.sqft_above=house_data.sqft_above.astype('int')
house_data.sqft_lot=house_data.sqft_lot.astype('int')


### TEST ###
house_data.info()
```

```python
house_data['bedrooms_zscore'] = zscore(house_data['bedrooms'])
bed_outliers=house_data.query('bedrooms_zscore > 3 or bedrooms_zscore < -3').index
bed_outliers
```

```python
### DEFINE ###
# delete the bedroom number ouside of the range 

# ### CLEAN ### 
house_data.bedrooms = house_data.bedrooms.drop(bed_outliers)

# ### TEST ###
house_data.describe()
```

```python
house_data.bedrooms
```

```python
house_data['bathrooms_zscore'] = zscore(house_data['bathrooms'])
bath_outliers=house_data.query('bathrooms_zscore > 3 or bathrooms_zscore < -3')
bath_outliers.sort_values(by='bathrooms',ascending=False)
```

```python
bath_to_drop=bath_outliers.query('bathrooms > bedrooms ').index
bath_to_drop
```

```python
### DEFINE ###
# delete the bathroom outliers that have bath # greater than bed # 

# ### CLEAN ### 
house_data.bathrooms=house_data.bathrooms.drop(bath_to_drop)

# ### TEST ###
house_data.describe()
```

```python
house_data['basement_zscore'] = zscore(house_data['sqft_basement'])
basement_outliers=house_data.query('basement_zscore > 3 or basement_zscore < -3')
basement_outliers.sort_values(by='sqft_basement',ascending=False)
#left in since they seemed resonable all are less than the sqft of living 
```

```python
house_data.describe()
```

```python
basement_outliers.sqft_basement.describe()
```

```python
house_data.sqft_living.describe()
```

```python
house_data['sqft_living_zscore'] = zscore(house_data['sqft_living'])
living_outliers=house_data.query('sqft_living_zscore > 3 or sqft_living_zscore < -3')
living_outliers.sort_values(by='sqft_living',ascending=False)
#kept the outliers since they seem resonable to me (its just a bigger house)
```

```python
living_outliers.sqft_living.describe()
```

```python
house_data['sqft_lot_zscore'] = zscore(house_data['sqft_lot'])
lot_outliers=house_data.query('sqft_lot_zscore > 3 or sqft_lot_zscore < -3')
lot_to_drop=lot_outliers.index
lot_outliers.sort_values(by='sqft_lot',ascending=False)
```
