### Step 2: Treat Names / Casing / Spelling Issues

```python
######################### TO DO ###################################

####### FOLLOW PROJECT DIRECTIONS TO COMPLETE THIS STEP ###########
```

```python
### DEFINE ###
#changing house_data column names to lowercase snake case.

### CLEAN ### 
house_data.rename(columns={'SQFT living':'Sqft_living'},inplace=True)
house_data.rename(columns={'SQFT lot':'Sqft_Lot'},inplace=True)
house_data.rename(columns={'waterfront?':'waterfront'},inplace=True)
house_data.rename(columns={'SQFT above':'Sqft_Above'},inplace=True)
house_data.rename(columns={'House Condition':'House_Condition'},inplace=True)
house_data.rename(columns={'SQFT basement':'Sqft_Basement'},inplace=True)
house_data.rename(columns={'Year Built':'Year_Built'},inplace=True)
house_data.rename(columns={'Year Renovated':'Year_Renovated'},inplace=True)
house_data.rename(columns={'SQFT living 15':'Sqft_Living_15'},inplace=True)
house_data.rename(columns={'SQFT lot 15':'Sqft_Lot_15'},inplace=True)
house_data.columns=house_data.columns.str.lower()

### TEST ###
house_data.head()
```

```python
### DEFINE ###
#fixing sales_data column names to lowercase snake case.

### CLEAN ### 
sales_data.columns=sales_data.columns.str.lower()

### TEST ###

sales_data.head()
```

### Step 3: Treat Missing Values

```python
######################### TO DO ###################################

####### FOLLOW PROJECT DIRECTIONS TO COMPLETE THIS STEP ###########
```

```python
### DEFINE ###
#drop unnamed: 0 column from house_data

### CLEAN ### 
house_data.drop(columns='unnamed: 0',inplace=True)

### TEST ###
house_data.head()
```

```python
### DEFINE ###
#drop unnamed: 0 column from sales_data

### CLEAN ### 
sales_data.drop(columns='unnamed: 0',inplace=True)

### TEST ###
sales_data.head()
```

```python
### DEFINE ###
#dropping the year_renovated column

### CLEAN ### 
house_data.isna().sum()
house_data.drop(columns='year_renovated',inplace=True)

### TEST ###
house_data.head()
```

```python
### DEFINE ###
# changing price to float and removing ',' and '$'
#sales id type change to string 
### CLEAN ### 
sales_data.price=sales_data.price.str.replace('$','')
sales_data.price=sales_data.price.str.replace(',','')
sales_data.price=sales_data.price.astype('float')
sales_data.id=sales_data.id.astype('string')
sales_data.date=pd.to_datetime(sales_data.date,"%d/%m/%Y")
### TEST ###
sales_data.head()
```

```python
### DEFINE ###
#since we only have a few (less than 10) null values we can just drop them.

### CLEAN ### 
house_data.dropna(inplace=True)
### TEST ###
house_data.info()
```

```python
### DEFINE ###
#house_data datatype fix

### CLEAN ### 
house_data.grade=house_data.grade.fillna(house_data.grade.mode()[0])
house_data.grade=pd.to_numeric(house_data.grade,errors='coerce').astype('int')
house_data.year_built=pd.to_numeric(house_data.year_built,errors='coerce').astype(int)

house_data.id=house_data.id.astype('string')
house_data.zipcode=house_data.zipcode.astype('string')
house_data.long=house_data.long.astype('string')
house_data.lat=house_data.lat.astype('string')
house_data.floors=house_data.floors.astype('int')
house_data.grade=house_data.grade.astype('int')
house_data.sqft_above=house_data.sqft_above.astype('int')
house_data.sqft_lot=house_data.sqft_lot.astype('int')


### TEST ###
house_data.info()
```
